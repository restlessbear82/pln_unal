```python
#Modificación de tablas. Lectura de nueva tabla denominada "elections"

ruta2 = '/content/gdrive/MyDrive/Colab Notebooks/Datos/elections.xlsx'

elections = pd.read_excel(ruta2)
elections.head(5)
```

```python
#Organizar los datos de acuerdo a una columna determinada, de mayor a menor (descendentemente)
elections.sort_values(["democrat"], ascending=False)
```

```python
#Total de votos por año, usando groupby y la instrucción sum
elections.groupby('year')[['democrat', 'republic']].sum()
```

```python
#Métodos de agrupamiento y organización de mayor a menor en un mismo código
elections_sort=(elections.groupby('year')[['democrat', 'republic']]
                .sum())
elections_sort['total'] = elections_sort['democrat'] + elections_sort['republic']
elections_sort = elections_sort.sort_values(['total'], ascending=False)
display(elections_sort)
```

```python
#Función pivot para reestructuración del dataframe

pivot_elections=elections.pivot(index='codecounty',
                columns='year',
                values=['democrat','republic','other'])
pivot_elections.head()
```

```python
#Dataframe con múltiples variables como índice
counties_multi=pd.read_excel(ruta, index_col=[0, 1])
display(counties_multi)
```

```python
#Suma de población de cada estado empleando el índice múltiple
counties_multi['population'].sum(level = 'codestate').head(5)
```

```python
#Concatenación de dataframes
#Creación de dataframe con los resultados de las elecciones del 2012

elections_2012 = elections[elections['year'] == 2012]
display(elections_2012)
```

```python
#Creación de un dataframe con los resultados de las elecciones del 2016, y una variable dummy que servirá como ejemplo
elections_2016 = elections[elections['year'] == 2016]
elections_2016['dummy'] = '0'
display(elections_2016)
```

```python
#Concatenación horizontal de dataframes mediante la instrucción concat

#Concatenación de dataframes 2012 y 2016. En el caso de 2012, al no tener valores de dummy, se llena con NaN (dato nulo o vacío)

elections_12_16=pd.concat([elections_2012, elections_2016])
display(elections_12_16)
```

```python
#Concatenación vertical de dataframes mediante la instrucción merge

#inner_join - Las tablas se unen mediante una columna en común, en este caso codecounty, y 
#solo deja los registros que tienen datos no nulos en ambas tablas
inner_joined=pd.merge(elections, counties)
inner_joined.head(10)
```

```python
#Outer join: une todas las columnas de las tablas, agregando los registros que tienen datos nulos
outer_joined=pd.merge(elections, counties, how='outer')
outer_joined.head(10)
```

```python
#Left join: une las dos tablas, dejando la totalidad de los registros de la primera tabla (la de la "izquierda")

left_joined=pd.merge(elections, counties, how='left')
left_joined.head(10)
```

```python
#Right join: une las dos tablas, dejando todos los registros de la segunda tabla (la de la "derecha")

right_joined=pd.merge(elections, counties, how='right')
right_joined.head(10)
```

```python
#Series de tiempo. Lectura de un nuevo dataframe en Pandas

ruta3 = '/content/gdrive/MyDrive/Colab Notebooks/Datos/airrpm.txt'

airrpm=pd.read_csv(ruta3, header=None, delimiter= '\s+', decimal=",")
airrpm.columns=["Time", "R1", "R2", "R3"]
airrpm.head()
```

```python
#Conversión de la variable 'Time' de tipo String a tipo 'Datetime' (tiempo), asignándola como índice y eliminando posteriormente Time

airrpm['Time']=pd.to_datetime(airrpm['Time'], format='%b-%y')
airrpm.index=airrpm['Time']
airrpm=airrpm.drop(['Time'], axis=1)
airrpm.head()
```

```python
#Obtención de estadísticas básicas de las variables al asignar el tiempo como índice
airrpm.groupby([airrpm.index.month])['R3'].describe()
```

```python
#Agrupación por año y suma de los valores para cada variable
airrpm.groupby([airrpm.index.year]).sum().head(10)
```
